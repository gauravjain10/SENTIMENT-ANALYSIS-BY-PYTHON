```python
from bs4 import BeautifulSoup
import urllib.request,re,time
import sys,nltk
from nltk.corpus import stopwords
import pandas as pd
import numpy as np


u1="http://www.amazon.in/Kindle-Glare-Free-Touchscreen-Display-Wi-Fi/product-reviews/B00KDRQ2RU/ref=undefined_2?showViewpoints=1&filterByStar=positive&pageNumber="
u2="http://www.amazon.in/Kindle-Glare-Free-Touchscreen-Display-Wi-Fi/product-reviews/B00KDRQ2RU/ref=undefined_2?showViewpoints=1&filterByStar=critical&pageNumber="
pos_rev=[]
crit_rev=[]

non_bmp_map = dict.fromkeys(range(0x10000, sys.maxunicode + 1), 0xfffd)


for page in range(1,40):
        url=u1+str(page)
        content = urllib.request.urlopen(url).read()
        soup = BeautifulSoup(content, "html.parser")
        
        for row in soup.find_all('span',attrs={"class" : "review-text"}):
            review=row.text.translate(non_bmp_map)
            pos_rev.append( review )
            time.sleep(1)
            
            
for page in range(1,40):
        url=u2+str(page)
        content = urllib.request.urlopen(url).read()
        soup = BeautifulSoup(content, "html.parser")
        
        for row in soup.find_all('span',attrs={"class" : "review-text"}):
            review=row.text.translate(non_bmp_map)
            crit_rev.append( review )
            time.sleep(1)      
            
pos=['Positive']*len(pos_rev)
crit=['Critical']*len(crit_rev)
all_rev=pos_rev+crit_rev
all_senti=pos+crit
d={'Review':all_rev,
   'Sentiment':all_senti }
labelled_data=pd.DataFrame(d,columns=['Review','Sentiment'])
```
